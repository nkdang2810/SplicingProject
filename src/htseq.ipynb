```python
name = "featureCounts"  
out_dir = Path(f"/rsrch4/home/mol_cgenesis/EMC_BIC_rsrch4/nkdang/Splicing_Projects/new/{name}")
out_dir.mkdir(parents=True, exist_ok=True)

log_dir = Path(
    f"/rsrch4/home/mol_cgenesis/EMC_BIC_rsrch4/nkdang/Splicing_Projects/new/log/{name}")
log_dir.mkdir(parents=True, exist_ok=True)

path_lsf = Path(
    f"/rsrch4/home/mol_cgenesis/EMC_BIC_rsrch4/nkdang/Splicing_Projects/new/lsf/{name}")
path_lsf.mkdir(parents=True, exist_ok=True)

file_mouse = ""
file_human = ""
for file in Path("/rsrch4/home/mol_cgenesis/EMC_BIC_rsrch4/nkdang/Splicing_Projects/new/data/alignments_clean_TOPHAT").glob("*/accepted_hits.bam"):
    filename = file.parent.name   
    if filename.startswith("2") or filename.startswith("3"):
        file_mouse += " " + str(file)
    else:
        file_human += " " + str(file)
command = f"#BSUB -W 24:00\n" \
                    f"#BSUB -q medium\n" \
                    f"#BSUB -n 12\n" \
                    f"#BSUB -M 128\n" \
                    f"#BSUB -R rusage[mem=128]\n" \
                    "#BSUB –u nkdang@mdanderson.org\n" \
                    f"#BSUB -J featureCounts_human\n" \
                    f"#BSUB -o {log_dir.joinpath('featureCounts_human.log')}\n" \
                    f"#BSUB –cwd /rsrch4/home/mol_cgenesis/EMC_BIC_rsrch4/nkdang/Splicing_Projects\n" \
                    f"featureCounts -M --fraction -s 2 -T 4 -p -t exon -g gene_id -a ref/gencode.v42.primary_assembly.annotation.gtf -o {out_dir.joinpath('humans.csv')}{file_human}"
with open(path_lsf.joinpath(f'humans.lsf'), "w+") as out:
    out.writelines(command)

command = f"bsub < {path_lsf.joinpath(f'humans.lsf')}"
subprocess.Popen(command, shell=True)

command = f"#BSUB -W 24:00\n" \
            f"#BSUB -q medium\n" \
            f"#BSUB -n 12\n" \
            f"#BSUB -M 128\n" \
            f"#BSUB -R rusage[mem=128]\n" \
            "#BSUB –u nkdang@mdanderson.org\n" \
            f"#BSUB -J featureCounts_mouse\n" \
            f"#BSUB -o {log_dir.joinpath('featureCounts_mouse.log')}\n" \
            f"#BSUB –cwd /rsrch4/home/mol_cgenesis/EMC_BIC_rsrch4/nkdang/Splicing_Projects\n" \
            f"featureCounts -M --fraction -s 2 -T 4 -p -t exon -g gene_id -a ref/gencode.vM31.primary_assembly.annotation.gtf -o {out_dir.joinpath('mouses.csv')}{file_mouse}"
with open(path_lsf.joinpath(f'mouses.lsf'), "w+") as out:
    out.writelines(command)

command = f"bsub < {path_lsf.joinpath(f'mouses.lsf')}"
subprocess.Popen(command, shell=True)
```

```python
results = pd.DataFrame()
for file in Path("/rsrch4/home/mol_cgenesis/EMC_BIC_rsrch4/nkdang/Splicing_Projects/new/htseq").glob("*"):
    # print(file)
    if file.name=="humans.count" or file.name=="mouses.count":
        continue
    if not (file.name.startswith("2") or file.name.startswith("3")):
        df = pd.read_csv(file, sep='\t', header=None, names=['gene',file.name.split(".")[0]]).iloc[:-5,:]
        if len(results)==0:
            results = df
        else:
            results = results.merge(df, on='gene', how='outer')
results = results.fillna(0).set_index('gene')
results = results.astype(int)
results = results.reindex(sorted(results.columns), axis=1)
results.to_csv("humans.count")
```
